```python
import pandas as pd 
```

```python
import os
```

```python
source = os.getcwd() # dataset root dir
a_paths = [] # list of paths to activities 'a'
for i in range(1,20):
    fname = 'a' + str(0) + str(i) if i<10 else 'a' + str(i)
    a_paths.append(os.path.join(source, fname))
print('a_paths contains:\n\t{}'.format(a_paths))
```

```python
p_fnames = [] # list of p files name [p1..p8]
for i in range(1,9):
    p_fnames.append('p'+str(i))
print('p_fnames contains:\n\t{}'.format(p_fnames))
```

```python
s_fnames = [] # list of s files name [s01..s60]
for i in range(1, 61): 
    n = 's' + str(0) + str(i) + '.txt' if i < 10 else 's' + str(i) + '.txt'
    s_fnames.append(n)
print('s_fnames size is {}'.format(len(s_fnames)))
print(s_fnames[0])
```

* **a_paths** contains paths to each Activity 

* **p_fnames** contains names of Subject folders 

* **s_fnames** contains names of Segments folders 

```python
headers = []
units = ['T', 'RA', 'LA', 'RL', 'LL']
dims = ['x', 'y', 'z']
sensors = ['acc', 'gyro', 'mag']
for unit in units:
    for sensor in sensors:
        for dim in dims:
            s = unit + '_' + dim + sensor
            headers.append(s)
print(headers)
```

```python
a_dfs = [] # list of activities df
p_dfs = [] # list of temp subject df
s_dfs = [] # list of temp segments df

for a in a_paths:
    for p in p_fnames:
        ppath = os.path.join(a, p) # construct a path to p
        for s in s_fnames: # read and prepare the segment files in p
            spath = os.path.join(ppath, s) # construct a path to a
            df = pd.read_csv(spath, delimiter=',', names=headers)
            df.insert(0, 'Segment', s[:3]) # add Segment column with s## as a value
            s_dfs.append(df)
        
        df = pd.concat(s_dfs) # merge all Segments that belongs to a Subject
        df.insert(0, 'Subject', p)
        p_dfs.append(df) # append to p_dfs
        s_dfs.clear() # empty s_dfs 
        
    df = pd.concat(p_dfs) # merge all Subjects that belongs to the an Activity
    df.insert(0, 'Activity', a[len(a)-3:len(a)])
    a_dfs.append(df) # append to p_dfs
    p_dfs.clear() # empty p_dfs 
len(a_dfs)
```

```python
dataset = pd.concat(a_dfs) # merge all activities data
```
